```python
import pandas as pd

# Load and preview the data
connection = connect()
monthly_spend_df = load_monthly_spend(connection)

# Drop rows without a year or month, ensure the date and month are numeric and create a datetime column from them
monthly_spend_df = monthly_spend_df.dropna(subset=["Year", "Month"])
monthly_spend_df["Year"] = monthly_spend_df["Year"].astype(int)
monthly_spend_df["Month"] = monthly_spend_df["Month"].astype(int)
monthly_spend_df["Date"] = pd.to_datetime(
    monthly_spend_df["Year"].astype(str) + "-" + monthly_spend_df["Month"].astype(str) + "-01"
)

# Sort by the new date column and take the last N months only
monthly_spend_df = monthly_spend_df.sort_values("Date").tail(NUMBER_OF_MONTHS)

display(monthly_spend_df)
```

```python
import numpy as np
import matplotlib.pyplot as plt

monthly_spend_df["label"] = monthly_spend_df["Date"].dt.strftime("%Y-%m")

# Trend line (same as before)
x = np.arange(len(monthly_spend_df))
trend = np.polyval(np.polyfit(x, monthly_spend_df["Spend"], 1), x)

plt.figure(figsize=(12, 5))
plt.bar(monthly_spend_df["label"], monthly_spend_df["Spend"])
plt.plot(monthly_spend_df["label"], trend, linestyle="--", color='orange')

plt.xticks(rotation=45)
plt.ylabel("Spend")
plt.title("Spend Over Time")

# Export the chart
export_chart("monthly-spend", "", "png")

plt.show()
```
